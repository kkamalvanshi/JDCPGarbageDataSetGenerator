```python
import pandas as pd
import random
from random import randrange
import datetime
```

```python
import string
```

```python
Software_ID = pd.DataFrame(columns = ['component', 'date_received', 'date_requested', 'date_programmed', 'software_assembly_id', 'software_assembly_id_upgrade', 'software_upgrade_date'])
```

```python
for i in range(10000):
    
    #component
    components = ['Component1', 'Component2', 'Component3', 'Component4']
    component = random.choice(components)
    
    #date_received
    date_received = datetime.datetime(randrange(2019,2021), randrange(1, 12), randrange(1, 25), randrange(0, 12), randrange(0, 59), randrange(0,59), randrange(1000000))

    #date_requested
    date_requested = datetime.datetime(randrange(2019,2021), randrange(1, 12), randrange(1, 25), randrange(0, 12), randrange(0, 59), randrange(0,59), randrange(1000000))

    
    #date_programmed
    date_programmed = datetime.datetime(randrange(2019,2021), randrange(1, 12), randrange(1, 25), randrange(0, 12), randrange(0, 59), randrange(0,59), randrange(1000000))

    #software_assembly_id
    letter_lst = list(string.ascii_uppercase)
    first_letter = random.choice(letter_lst)
    first_num = str(randrange(0,9))
    lst = [first_letter, first_num, 
          str(random.choice(letter_lst)), str(randrange(0,9)),
          str(random.choice(letter_lst)), str(randrange(0,9))]
    software_assembly_id = ''.join(lst)
    
    
    #software_assembly_id_upgrade
    
    lst = [first_letter, first_num, 
          str(random.choice(letter_lst)), str(randrange(0,9)),
          str(random.choice(letter_lst)), str(randrange(0,9))]
    software_assembly_id_upgrade = ''.join(lst)
    
    
    #software_upgrade_date
    software_upgrade_date = datetime.datetime(randrange(2019,2021), randrange(1, 12), randrange(1, 25), randrange(0, 12), randrange(0, 59), randrange(0,59), randrange(1000000))


    new_row = {'component' : component, 
           'date_received' : date_received, 
           'date_requested' : date_requested, 
           'date_programmed' : date_programmed, 
           'software_assembly_id' : software_assembly_id, 
           'software_assembly_id_upgrade' : software_assembly_id_upgrade, 
           'software_upgrade_date' : software_upgrade_date}

    Software_ID = Software_ID.append(new_row, ignore_index = True)
```
