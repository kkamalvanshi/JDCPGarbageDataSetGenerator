```python
import pandas as pd
import random
from random import randrange
import datetime
```

```python
Measurements = pd.DataFrame(columns = ['cnfgr_msrmt_id', 'elmt_seq_num', 'beg_cptr_ts', 'end_cptr_ts', 
                                    'msrmt_duration_in_min', 'val', 'val_cnt', 'msrmt_typ_dsc', 'agrgtn_typ_dsc',
                                    'msrmt_dsc', 'dim_dsc', 'uom', 'elmt_dsc', 'msrmt_nm'])
```

```python
for i in range(10000):
    
    #cnfgr_msrmt_id
    cnfgr_msrmt_id = randrange(1000)
    
    #elmt_seq_num
    elmt_seq_num = randrange(3)
    
    #beg_cptr_ts
    
    year = randrange(2019,2021)
    month = randrange(1, 12)
    day = randrange(1, 25)
    beg_cptr_ts = datetime.datetime(year, month, day, randrange(0, 12), randrange(0, 59), randrange(0,59), randrange(1000000))

    #end_cptr_ts
    end_cptr_ts = datetime.datetime(year, month, day, randrange(0, 12), randrange(0, 59), randrange(0,59), randrange(1000000))

    #msrmt_duration_in_min
    msrmt_duration_in_min = (end_cptr_ts - beg_cptr_ts).seconds/60
    
    #val
    val = msrmt_duration_in_min/60
    
    #val_cnt
    val_cnt = randrange(20)
    
    #msrmt_typ_dsc
    msrmt_typ_dscs = ['timer', 'min val', 'max val', 'last', 'integral']
    msrmt_typ_dsc = random.choice(msrmt_typ_dscs)
    
    #agrgtn_typ_dsc
    agrgtn_typ_dsc = randrange(10)
    
    #msrmt_dsc
    msrmt_dsc = randrange(10)
    
    #dim_dsc
    dim_dscs = ['bucket1', 'bucket2', 'bucket3', 'bucket4', 'bucket5']
    dim_dsc = random.choice(msrmt_typ_dscs)
    
    #uom
    uoms = ['seconds', 'KPH']
    uom = random.choice(uoms)
    
    #elmt_dsc
    elmt_dsc = randrange(10)
    
    #msrmt_nm
    msrmt_nms = ['bucket1', 'bucket2', 'bucket3', 'bucket4', 'bucket5']
    msrmt_nm = random.choice(msrmt_nms)

    new_row = {'cnfgr_msrmt_id' : cnfgr_msrmt_id, 
           'elmt_seq_num' : elmt_seq_num, 
           'beg_cptr_ts' : beg_cptr_ts, 
           'end_cptr_ts' : end_cptr_ts, 
           'msrmt_duration_in_min' : msrmt_duration_in_min, 
           'val' : val, 
           'val_cnt' : val_cnt, 
           'msrmt_typ_dsc' : msrmt_typ_dsc, 
           'agrgtn_typ_dsc' : agrgtn_typ_dsc, 
           'msrmt_dsc' : msrmt_dsc, 
           'dim_dsc' : dim_dsc, 
           'uom' : uom, 
           'elmt_dsc' : elmt_dsc, 
           'msrmt_nm' : msrmt_nm}

    Measurements = Measurements.append(new_row, ignore_index = True)
```
