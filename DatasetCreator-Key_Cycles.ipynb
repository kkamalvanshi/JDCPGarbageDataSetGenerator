```python
import pandas as pd
import random
from random import randrange
import datetime
```

```python
Key_Cycles = pd.DataFrame(columns = ['key_cycle_id', 'key_cycle_start', 'key_cycle_end', 'key_on_hrs', 'key_cycle_engine_hrs', 'hours_since_prev_key_cycle'])
```

```python
for i in range(1000):
    
    #key_cycle_id
    key_cycle_id = randrange(1000000, 100000000)
    
    #key_cycle_start
    year = randrange(2019,2021)
    month = randrange(1, 12)
    day = randrange(1, 25)
    key_cycle_start = datetime.datetime(year, month, day, randrange(0, 12), randrange(0, 59), randrange(0,59), randrange(1000000))

    #key_cycle_end
    key_cycle_end = datetime.datetime(year, month, day, randrange(0, 12), randrange(0, 59), randrange(0,59), randrange(1000000))
    
    #key_on_hrs
    key_on_hrs = (key_cycle_end - key_cycle_start).seconds/3600
    

    #key_cycle_engine_hrs
    key_cycle_engine_hrs = random.uniform(0, 24)
    
    #hours_since_prev_key_cycle
    hours_since_prev_key_cycle = random.uniform(0, 24)


    new_row = {'key_cycle_id' : key_cycle_id, 
           'key_cycle_start' : key_cycle_start, 
           'key_cycle_end' : key_cycle_end, 
           'key_on_hrs' : key_on_hrs, 
           'key_cycle_engine_hrs' : key_cycle_engine_hrs, 
           'hours_since_prev_key_cycle' : hours_since_prev_key_cycle}

    Key_Cycles = Key_Cycles.append(new_row, ignore_index = True)
```
